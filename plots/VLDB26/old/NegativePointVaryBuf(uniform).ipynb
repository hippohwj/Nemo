```python
# Line plot version of PhantomProtectionVaryBufWhenSkew (complete cell)

import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from process_helper import select_color_idx

# style
txtsize = 30
ticksize = 25
line_width = 4
ms = 10

x_dimension = "g_total_buf_sz"
_, line_colors = select_color_idx(4)
nemo_batch_evict_c = "#31a354"
nemo_record_evict_c = line_colors["BLUE"]
pcl_c = line_colors["ORANGE"]

nemo_batch_evict_label = "NemoSentinel"
nemo_record_evict_label = "NemoNK"
pcl_label = "PCL(10M)"

# data
datadir = "../../output/phantom_protection/"
exp_tuple_batch_evict = "pp_varybuf_d10g_skewscatter"
exp_tuple_record_evict = "pp_varybuf_d10g_skewscatter"
exp_pcl = "pp_varybuf_d10g_skewscatter"

data_tuple_batch = [json.loads(line) for line in open(datadir + exp_tuple_batch_evict + ".txt")]
data_tuple_batch = [item for item in data_tuple_batch if item["g_phantom_protection_type"] == "PHANTOM_NEXT"]
df_tuple_batch = pd.DataFrame.from_records(data_tuple_batch).sort_values(by=x_dimension)

data_tuple_record = [json.loads(line.replace("inf", "0")) for line in open(datadir + exp_tuple_record_evict + ".txt")]
data_tuple_record = [item for item in data_tuple_record if item["g_phantom_protection_type"] == "NEXT_KEY"]
df_tuple_record = pd.DataFrame.from_records(data_tuple_record).sort_values(by=x_dimension)

data_pcl = [json.loads(line.replace("inf", "0")) for line in open(datadir + exp_pcl + ".txt")]
data_pcl = [item for item in data_pcl if item["g_phantom_protection_type"] == "PCL"]
df_pcl = pd.DataFrame.from_records(data_pcl).sort_values(by=x_dimension)

# x = buffer ratio
x_nemo = df_tuple_batch[x_dimension] / (df_tuple_batch['g_record_size'] * df_tuple_batch['num_rows_'])
x_rc   = df_tuple_record[x_dimension] / (df_tuple_record['g_record_size'] * df_tuple_record['num_rows_'])
x_pcl  = df_pcl[x_dimension] / (df_pcl['g_record_size'] * df_pcl['num_rows_'])

# # y = aborts per insert txn
# y_nemo = df_tuple_batch["rw_abort_cnt_"] / df_tuple_batch["insert_commit_cnt_"]
# y_rc   = df_tuple_record["rw_abort_cnt_"] / df_tuple_record["insert_commit_cnt_"]
# y_pcl  = df_pcl["rw_abort_cnt_"] / df_pcl["insert_commit_cnt_"]


# replace the y computations
def abort_rate(df):
    a = df["rw_abort_cnt_"].astype(float)
    c = df["insert_commit_cnt_"].astype(float)
    return a / (a + c)

y_nemo = abort_rate(df_tuple_batch)
y_rc   = abort_rate(df_tuple_record)
y_pcl  = abort_rate(df_pcl)

# axis settings



# figure
fig, ax1 = plt.subplots(figsize=(10, 8))

ax1.plot(x_nemo, y_nemo, color=nemo_batch_evict_c, marker="s", markersize=ms, linewidth=line_width, label=nemo_batch_evict_label)
ax1.plot(x_rc,   y_rc,   color=nemo_record_evict_c, marker="^", markersize=ms, linewidth=line_width, label=nemo_record_evict_label)
ax1.plot(x_pcl,  y_pcl,  color=pcl_c,               marker="o", markersize=ms, linewidth=line_width, label=pcl_label)

# ticks as percentages
ticks = np.unique(np.concatenate([x_nemo.to_numpy(), x_rc.to_numpy(), x_pcl.to_numpy()]))
# ax1.set_xticks([0.1, 0.3, 0.5, 0.7, 0.9])
ax1.set_xticks(ticks)
labels = [f'{v*100:.2f}%' if v < 0.01 else f'{v*100:.0f}%' for v in ticks]
# lables = ['10%', '30%', '50%', '70%', '90%']
ax1.set_xticklabels(labels, fontsize=ticksize)

ax1.tick_params(axis='y', labelsize=ticksize)
# ax1.set_yscale('log')
ax1.set_xlim(0, 1)
ax1.set_xlabel("Buffer Ratio(Buffer Size/Data Size)", fontsize=txtsize)

ax1.set_ylabel("Abort Rate", fontsize=txtsize)
# ax1.set_yscale("linear")
ax1.set_ylim(0, 1.1)

# ax1.set_ylabel("Avg Aborts per Txn (Insert Txn)", fontsize=txtsize)
ax1.grid(axis='y', linestyle='--', linewidth=0.25)

ax1.legend(fontsize=ticksize, bbox_to_anchor=(-0.01, 1, 0.9, .102), loc=2,
           ncol=3, frameon=False, columnspacing=0.4, handletextpad=0.1)

plt.show()
```
